## Chunking

### Fixed Length Splitting

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=0,separators=["\n\n", "\n", ".", " ", ""])

# I will give a random text, first we will analyze the text length and then we can chunk it using the TextSplitter()

#text about Batman in 200 words

Batman_text = """Beneath the jagged skyline of Gotham City, a transformation occurs as the sun dips below the horizon. The silhouette of a bat against the moon is not merely a signal; it is a promise of reckoning. Born from the trauma of a shattered childhood, Bruce Wayne mission is a relentless pursuit of an impossible ideal: a city where no child has to lose their parents to the darkness of an alleyway.

Unlike the gods and aliens who inhabit his world, Batmans power lies in the limits of human potential. He is a master of a dozen disciplines—from forensic science and escapology to every known martial art—honed through decades of grueling discipline. His utility belt holds the gadgets of a billionaire, but his true strength is a tactical mind that stays three steps ahead of both allies and enemies.

He navigates a rogues' gallery that reflects the fractured psyche of Gotham itself, standing as the thin line between order and total collapse. To the criminals, he is a wraith; to the citizens, he is a myth; but to himself, he is a necessary shadow. Cloaked in his cape, the Caped Crusader remains the worlds greatest detective, a silent sentinel who sacrifices his own peace to ensure Gothams survival."""

print(f'The length of the text for chunking : {len(Batman_text)}')

chunks = text_splitter.split_text(Batman_text)
print(f"Total chunks : {len(chunks)}\n")



for chunk in chunks:
    print(chunk)




```

### Chunking a Document (realtime utility)
 - first step is we will load a document using document loaders 
 - second step is we chunk that document using document splitter
   - so firstly we will load a wikipedia webpage.
   - later on we will load and chunk a actual pdf.

#### Document Loading from web based docuemnt loader of Langchain

```python
from langchain_community.document_loaders import WebBaseLoader

url = "https://en.wikipedia.org/wiki/Batman"

webpageloader = WebBaseLoader(url)

#checking what type the webpageloader is , just for curiosity

print(type(webpageloader))

web_docs = webpageloader.load()
print(type(web_docs))
print(web_docs)

webpage_content = web_docs[0].page_content
print(type(webpage_content))
print(webpage_content)
print(len(webpage_content))
```

#### PDF loading through PDF loader of Langchain using PyPDFLoader

```python
from langchain_community.document_loaders import PyPDFLoader

pdf_path = "./resources/Panchatantra.pdf"

pdf_loader = PyPDFLoader(pdf_path)

print(type(pdf_loader))

pdf_data = pdf_loader.load()

print(pdf_data[0])
```

### Now we will chunk both of them , the Web Doc and the PDF

- Webpage chunking

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

doc_chunker = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
#chunk_overlap is provided minimum 5 to 10 % of the chunk size to preserve the context

webpage_chunks = doc_chunker.split_documents(web_docs)
print(f"Web page total chunks : {len(webpage_chunks)}")

print(type(webpage_chunks))

#list to embed

webpage_embedding_list = []

for chunk in webpage_chunks:
    webpage_embedding_list.append(chunk.page_content)

print(webpage_embedding_list)

type(webpage_embedding_list) #list

# for chunk in webpage_chunks:
#     print(chunk.page_content)
```

- PDF chunking

```python
pdf_data_chunks = doc_chunker.split_documents(pdf_data)

print(len(pdf_data_chunks))
```

## Once the chunking part is done we will be moving on to Embedding part
  

```python
from langchain_google_vertexai import VertexAIEmbeddings

embeddings = VertexAIEmbeddings(model="text-embedding-005", project="my-genai-playground", location="asia-south1")

sample_embedding = embeddings.embed_query(text="Hii this is you boy Bruice Wayne")

#since Google Vertex Ai the model we are using text-embedding-005 only allows us to use 20,000 chunks to embed at a time , we need to split our chunks that way so that we can embed under the limit.

# 1. Define your batch size (20-50 is usually safe for Vertex AI)
batch_size = 50
all_embeddings = []

# 2. Loop through your list in steps
for i in range(0, len(webpage_embedding_list), batch_size):
    # Grab a slice of your list
    batch = webpage_embedding_list[i : i + batch_size]
    
    print(f"Processing batch: {i} to {i + batch_size}...")
    
    # Send only this small batch to the API
    batch_embeddings = embeddings.embed_documents(batch)
    
    # Add the results to our final list
    all_embeddings.extend(batch_embeddings)

print(f"Success! Total vectors generated: {len(all_embeddings)}")


```

## Since we are now done with Embeddings now we will move on to Vector DB storage implementation.

- You can visualize the working of vector database using this vector visualizer website
- [projector.tensorflow.org](https://projector.tensorflow.org/)

#### Step 1 is to initialize the Vector database in our local system. Here we are using Chroma

```python
from langchain_chroma import Chroma

vector_store = Chroma(
    collection_name="webpage_rag_chunks",
    embedding_function=embeddings,
    persist_directory="./batman_chroma_langchain_db",  # Where to save data locally, remove if not necessary
)

```

```python
# vector_store.add_documents(documents=webpage_chunks)

# Using the same batching logic for inserting chunks into the chroma db
batch_size = 50

for i in range(0, len(webpage_chunks), batch_size):
    # Slice the list of Document objects
    chunk_batch = webpage_chunks[i : i + batch_size]
    
    print(f"Adding batch {i} to {i + batch_size} to the Vector DB...")
    
    # This call will now stay under the 20,000 token limit
    vector_store.add_documents(documents=chunk_batch)

print("Success! All Batman chunks have been embedded and stored.")
```

- since we have stored all the embeddings to our vector storage , now we can ask questions to it to checking its working.

```python
# 1. Define your question
query = "Who is Bruce Wayne's secret identity?"

# 2. Search the database
# k=2 tells Chroma to return the top 2 most relevant chunks
results = vector_store.similarity_search(query, k=2)

print(results) #results are in List of Lists , so it is difficult to read, let me make is more readable.

# 3. Print the results to verify
for i, doc in enumerate(results):
    print(f"\n--- Result {i+1} ---")
    print(f"Content: {doc.page_content[:300]}...") # Print first 300 chars
    print(f"Metadata: {doc.metadata}")

```
